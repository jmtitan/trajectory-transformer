```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
comparison_times = np.loadtxt(open("./comparison_times.csv","rb"), delimiter=",")
expand_times = np.loadtxt(open("./expand_times.csv","rb"), delimiter=",")
comparison_times.shape
```

```python
plt.plot(comparison_times[1:], label="Best-first BS")
plt.ylabel("Time/s")
plt.xlabel("Maximum Comparisons")
plt.grid()
plt.axhline(1.81,ls=":",c='red', label="BS")
plt.legend()
plt.savefig("comparison.png", dpi=800)
plt.show()
```

```python
import matplotlib

BIGGER_SIZE = 20
plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

plt.plot(expand_times[1:],lw=2, label="Best-first BS")
plt.ylabel("Time/s")
plt.xlabel("Exploration Branches")
plt.grid()

plt.axhline(1.81,ls=":", lw=4,c='red', label="BS")
plt.legend()


plt.savefig("expand.png", dpi=800, bbox_inches='tight')
plt.show()
```
